```python
import pandas as pd
import plotly.express as px
```

```python
# DataFrame do catálogo da Netflix na américa latina
tabela = pd.read_csv("/content/drive/MyDrive/Netflix-America-Latina-csv.csv", encoding="latin1")
tabela = tabela.drop(["show_id", "date_added", "title", "director", "duration"], axis=1)

display(tabela)
```

```python
# Quais paises mais aparecem na tabela por porcentagem.
qtde_paises = tabela["country"].value_counts()

# Eliminando variações.
delete_paises = ["Argentina, Chile", "Uruguay, Guatemala", "Paraguay, Argentina", "Chile, Peru", "Chile, Brazil","Colombia, Mexico", "Venezuela, Colombia", "Argentina, Chile, Peru", "Chile, Argentina", "Argentina, Venezuela", "Mexico, Brazil", "Mexico, Argentina", "Uruguay, Argentina", "Chile, Brazil"]

tabela = tabela[~tabela['country'].isin(delete_paises)] # ~ = inverte a seleção indicada, .isin = verifica cada valor na coluna e retorna True or False.
tabela = tabela.dropna()
df = qtde_paises
df.plot.pie( y='country', autopct='%1.1f%%', figsize=(8, 8), title='Paises Latinos com maior posição.')
```

```python
# Quais paises mais aparecem na tabela por quantidade.
scatter_plot = px.bar(tabela, y="country", color='country')
# Exibir o gráfico
scatter_plot.show()
```

```python
# Selecionando apenas as linhas onde o país é "Mexico"
tabela_mexico = tabela[tabela["country"] == "Mexico"]
# Contando os valores na coluna "rating" que contém o valor "Mexico"
rating_mexico = tabela_mexico["rating"].value_counts()

rating_mexico.plot.bar(figsize=(8, 8), title='Distribuição da faixa etária no México')
```

```python
# Número de produções por ano.
ano_prod = tabela["release_year"].value_counts().sort_index()
# Plotar o gráfico
ano_prod.plot.bar(figsize=(8, 8), title='Anos com mais produções.')
```

```python
# Número de produções por ano do méxico.
ano_prod_mexico = tabela_mexico["release_year"].value_counts().sort_index()
# Plotar o gráfico
ano_prod_mexico.plot.bar(figsize=(8, 8), title='Anos com mais produções do México.')
```

```python
# Filtrar os dados para o ano de 2016
df_ano = tabela_mexico[tabela_mexico["release_year"] == 2016]

# Contagem das categorias dos filmes
df_categoria = df_ano["listed_in"].value_counts()

bar_plot = px.bar(y=df_categoria.index, x=df_categoria.values, title="Categorias mais produzidas no México em 2016")
bar_plot.show()
```
